```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

$$ \Large {\displaystyle f(x)=3e^{-{\frac {x^{2}}{8}}}} \sin(x/3)$$

* Find the first four terms of the Taylor expansion of the above equation
* Make a plot of the function
* Plot size 10 in x 4 in
* X limts -5, 5
* Y limits -2, 2
* Over-plot the 1-term Taylor expansion using a different color
* Over-plot the 2-term Taylor expansion using a different color
* Over-plot the 3-term Taylor expansion using a different color
* Over-plot the 4-term Taylor expansion using a different color

```python
sp.init_printing()

x = sp.symbols('x')

my_x = np.linspace(-10,10,100)
```

```python
equation = 3 * sp.exp(-x**2 / 8) * sp.sin(x/3)
sp.series(equation, x, n=8)
```

```python
function_y = 3 * np.exp(-my_x**2 / 8) * np.sin(my_x/3)
first_y = my_x
second_y = first_y - (31/216) * my_x**3
third_y = second_y + (1591/155520) * my_x**5
forth_y = third_y - (113653/235146240) * my_x**7
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(10, 4)
fig.tight_layout()
ax.set_xlim(-5,5)
ax.set_ylim(-2,2)

ax.plot(my_x, function_y, color='b', marker='None', linewidth=3, linestyle='-');
ax.plot(my_x, first_y, color='r', marker='None', linestyle='-');
ax.plot(my_x, second_y, color='g', marker='None', linestyle='-');
ax.plot(my_x, third_y, color='c', marker='None', linestyle='-');
ax.plot(my_x, forth_y, color='y', marker='None', linestyle='-');
```
